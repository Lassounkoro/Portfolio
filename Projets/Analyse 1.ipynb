```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
df = pd.read_csv("données/2019.csv")

# Aperçu
print(df.head())

# Nettoyage
df = df.dropna()

# Renommage pour simplifier
df = df.rename(columns={
    "Score": "score",
    "GDP per capita": "gdp",
    "Healthy life expectancy": "life_expectancy",
    "Perceptions of corruption": "corruption",
    "Country or region": "country"
})

# Corrélations
corr = df[["score", "gdp", "life_expectancy", "corruption"]].corr()
print(corr)

# Visualisation corrélation
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Corrélation entre les indicateurs du bonheur - 2019")
plt.show()

# Top 10 pays les plus heureux
top10 = df.sort_values("score", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="country", y="score", data=top10)
plt.xticks(rotation=45)
plt.title("Top 10 pays les plus heureux - 2019")
plt.show()
```
